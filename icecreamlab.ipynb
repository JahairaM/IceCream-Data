```python
import pandas as pd
import seaborn as sns

#
data = pd.DataFrame({
    'Temperature (F)': [57.56, 61.52, 53.42, 59.36, 65.3, 71.78, 66.92, 77.18, 74.12, 64.58],
    'Sales': [215, 325, 185, 332, 406, 522, 412, 614, 544, 421]
})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Temperature (F)', y='Sales', color='b')
plt.title('Frozen Yogurt Sales vs Temperature')
plt.xlabel('Temperature (F)')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# Monthly data for total units
monthly_data = pd.DataFrame({
    'Month': list(range(1, 13)),
    'Total Units': [21100, 18330, 22470, 22270, 20960, 20140, 29550, 36140, 23400, 26670, 41280, 30020]
})

# Create line plot
plt.figure(figsize=(10, 6))
plt.plot(monthly_data['Month'], monthly_data['Total Units'], marker='o', color='b')
plt.title('Total Units Sold Across Months')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.xticks(range(1, 13))
plt.show()
```

```python
# Monthly data for total profit
monthly_profit_data = pd.DataFrame({
    'Month': list(range(1, 13)),
    'Total Profit': [211000, 183300, 224700, 222700, 209600, 201400, 295500, 361400, 234000, 266700, 412800, 300200]
})

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(monthly_profit_data['Month'], monthly_profit_data['Total Profit'], color='g')
plt.title('Total Profit Across Months')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.xticks(range(1, 13))
plt.show()
```
